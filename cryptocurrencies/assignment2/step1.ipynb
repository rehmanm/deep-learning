```python
#Import all the libraries
import pandas_datareader.data as web
import datetime
import matplotlib.pyplot as plt
```

```python
start = datetime.datetime(2018, 1, 1)
end = datetime.datetime(2018, 2, 27)
apple = web.DataReader('AAPL', 'google', start, end)
fb = web.DataReader('FB', 'google', start, end)
```

```python
plt.plot(apple.index, apple.High, color='r')
plt.plot(fb.index, fb.High, color='b')
plt.show()
```
